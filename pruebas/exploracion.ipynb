```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.load('../assets/proyecto_training_data.npy')
data
```

```python
n = len(data)
n80 = int(n * 0.8)
n80
```

```python
data_training = data[0:n80]
data_testing = data[n80:n]
data_training
```

```python
training_df = pd.DataFrame(data_training, columns=['SalePrice', 'OverallQual', '1stFlrSF', 'TotRmsAbvGrd', 'YearBuilt', 'LotFrontage'])
training_df
```

##Análisis exploratorio de datos

```python
# Medias
for col in training_df.columns:
    print(f'Media {col}: ', training_df[col].mean())
```

```python
# Valor máximo
for col in training_df.columns:
    print(f'Máximo {col}: ', training_df[col].max())
```

```python
# Valor mínimo
for col in training_df.columns:
    print(f'Mínimo {col}: ', training_df[col].min())
```

```python
# Rango
for col in training_df.columns:
    print(f'Rango {col}: ', training_df[col].max() - training_df[col].min())
```

```python
# Desviación Estandar
for col in training_df.columns:
    print(f'Desviación Estandar {col}: ', training_df[col].std())
```

##Histogramas de variables

```python
# axs = []
for col in training_df.columns:
    fig, axs = plt.subplots()
    sns.histplot(training_df[col], kde=True, ax=axs)
```
